```python
class Player:
    score = 10
    
    def updateScore(self, score):
        self.score = self.score + score    
    
player_1 = Player()    
print("fist match total score :", player_1.score)   
player_1.updateScore(30)
print("second match total score : ", player_1.score)
     
```
