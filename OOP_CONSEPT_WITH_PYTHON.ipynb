```python
# create a python class named circle that repersent the circle. 
# your class should be able to 
# 1.stor the radius of the circle 
# 2.calculate and return the area of the cicle
# 3.calculate and return the circumference of the circle 
# 4.iclude the method display or that print the radius, area and circomference of the circle in the clear rederble form
from math import pi

class circle:
    radius = 0
    
    def __init__(self , radius):
        self.radius = radius
        
    def Circale_Circomference(self):   
        return 2 * pi * self.radius
        
    def Circale_Area(self):
        return pi * self.radius * self.radius
    
    def display_info(self):
        return f"redius: {self.radius} , Circale_Circomference: {self.Circale_Circomference()} , Circale_Area: {self.Circale_Area()}"
        
circle_01 = circle(7)
print(circle_01.Circale_Circomference()) 
print(circle_01.Circale_Area())      
print(circle_01.display_info())
```

```python
class Animal:
    def __init__(self ,name):
        self.name = name
        
    def speak(self):
        print(f"{self.name} , speak")
        


class Dog(Animal):
    def __init__(self , name , color):
        self.name = name
        self.color = color
        
    def speak(self):
        print(f"{self.color} {self.name} , barks")

dog_1 = Dog("chidu" , "brown")

dog_1.speak()
```
